给定一个二叉树，返回它的中序 遍历。

示例:

输入: [1,null,2,3]

       1
        \
         2
        /
       3

输出: [1,3,2]

```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

        
# 法1：递归：
def inorderTraversal(root):
    res = []
    helper(root)
    return res

    def helper(root):
        if not root:
            return 
        helper(root.left)
        res.append(root.val)
        helper(root.right)
        
# 法2：栈
def inorderTraversal1(root):
    if not root:
        return []
    res = []
    stack = []
    p = root
    while stack or p:
        while p:
            stack.append(p)
            p = p.left
        top = stack.pop()
        res.append(top.val)
        p = top.right
    return res
```

```python
# 感觉下面这个思路更清晰一些
def preT(root):
    if not root:
        return []
    stack = [root]
    res = []
    while stack:
        top = stack.pop()
        res.append(top.val)
        if top.right:
            stack.append(top.right)
        if top.left:
            stack.append(top.left)
    return res




# 一种简洁的方法，根-右-左倒过来就是后根序，而根右左容易实现：
def postT(root):
    if not root:
        return []
    res = []
    stack = [root]
    while stack:
        t = stack.pop()
        res.append(t.val)
        if t.left:
            stack.append(t.left)
        if t.right:
            stack.append(t.right)
    return res[::-1]
            
```
