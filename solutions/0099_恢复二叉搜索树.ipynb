99. 恢复二叉搜索树

给你二叉搜索树的根节点 root ，该树中的 恰好 两个节点的值被错误地交换。请在不改变其结构的情况下，恢复这棵树 。

示例 1：
```
输入：root = [1,3,null,null,2]
输出：[3,1,null,null,2]
解释：3 不能是 1 的左孩子，因为 3 > 1 。交换 1 和 3 使二叉搜索树有效。
```
示例 2：
```
输入：root = [3,1,4,null,null,2]
输出：[2,1,4,null,null,3]
解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。
```

提示：
- 树上节点的数目在范围 [2, 1000] 内
- -231 <= Node.val <= 231 - 1

进阶：使用 O(n) 空间复杂度的解法很容易实现。你能想出一个只使用 O(1) 空间的解决方案吗？

````python
'''2022, M'''
from utils import *
class Solution:
    """ 显式中序遍历，空间n """
    def recoverTree(self, root: Optional[TreeNode]) -> None:
        """
        Do not return anything, modify root in-place instead.
        """
        # 想法：先中序遍历(递增序)，找到异常值；然后交换树的val，注意只是交换val，而不是内存
        swap_val = self.get_swap_node(root)
        self.recover(swap_val)

    def recover(self, swap_nodes):
        tmp_val = swap_nodes[0].val
        swap_nodes[0].val = swap_nodes[1].val
        swap_nodes[1].val = tmp_val
        
    def get_swap_node(self, root):
        node_list = [TreeNode(-float('inf'))]
        stack = []
        cur = root
        while stack or cur:
            while cur:
                stack.append(cur)
                cur = cur.left
            cur = stack.pop()
            node_list.append(cur)
            cur = cur.right
        node_list.append(TreeNode(float('inf')))  # 加个inf方便判断异常值

        def is_invalid(idx):
            if node_list[idx-1].val < node_list[idx].val < node_list[idx+1].val:
                return False
            else:
                return True

        swap_nodes = list()
        for i in range(1, len(node_list) - 1):
            if is_invalid(i):
                swap_nodes.append(node_list[i])
                print(node_list[i].val)
        return (swap_nodes[0], swap_nodes[-1])
    

class Solution1:
    """ 隐式中序遍历，空间H, 二叉树高度"""
    def recoverTree(self, root: Optional[TreeNode]) -> None:
        """
        Do not return anything, modify root in-place instead.
        """
        swap_val = self.get_swap_node(root)
        self.recover(swap_val)

    def recover(self, swap_nodes):
        tmp_val = swap_nodes[0].val
        swap_nodes[0].val = swap_nodes[1].val
        swap_nodes[1].val = tmp_val

    def get_swap_node(self, root):
        pre_node = None
        x, y = None, None
        stack = []
        cur = root
        while stack or cur:
            while cur:
                stack.append(cur)
                cur = cur.left
            cur = stack.pop()
            if pre_node and cur.val < pre_node.val:
                y = cur
                if not x:
                    x = pre_node
                    # print(x.val)
                else:
                    # print(y.val)
                    break
            pre_node = cur
            cur = cur.right
        return (x, y)
            


# root = l2tn([3, 1, 4, None, None, 2])
root = l2tn([1,3,None,None,2])
# Solution().recoverTree(root)
Solution1().recoverTree(root)
pprint(root.to_dict())
````
